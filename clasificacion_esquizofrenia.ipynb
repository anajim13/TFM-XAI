```python
from sklearn.model_selection import train_test_split
import time
train, test = train_test_split(df_esq, test_size = 0.30)

variables_train = train[['Hora5','Hora6','Hora7','Hora8','Hora9','Hora10','Hora11','Hora12','Hora13','Hora14','Hora15']].to_numpy() 
variables_test = test[['Hora5','Hora6','Hora7','Hora8','Hora9','Hora10','Hora11','Hora12','Hora13','Hora14','Hora15']].to_numpy()

tipo_train = train['Tipo'].to_numpy()
tipo_test = test['Tipo'].to_numpy()
```

```python
from sklearn.datasets import load_iris 
from sklearn.linear_model import LogisticRegression 
inicio_RL = time.time()
reg_log = LogisticRegression(random_state=0, max_iter=10000).fit(variables_train, tipo_train) 
fin_RL = time.time()
t_RL = fin_RL - inicio_RL 
print("El modelo tarda en procesarse",t_RL, "segundos.")
print("Se predicen un", reg_log.score(variables_train, tipo_train) * 100, "% de los casos con Regresión Logística.")

#REDES NEURONALES
from sklearn.neural_network import MLPClassifier
inicio_RN = time.time()
redes = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15, 25), random_state=1).fit(variables_train,tipo_train)
fin_RN = time.time()
t_RN = fin_RN - inicio_RN 
print("El modelo de Redes Neuronales tarda en procesarse",t_RN, "segundos.")
print("Se predicen un",redes.score(variables_train,tipo_train)* 100, "% de los casos con Redes Neuronales.")

#KERAS
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers.core import Dense
RL_Keras = Sequential()
RL_Keras.add(Dense(units=64, activation = 'relu'))
RL_Keras.add(Dense(units=32, activation = 'ELU'))
RL_Keras.add(Dense(units=16, activation = 'ELU'))
RL_Keras.add(Dense(units=1, activation = 'sigmoid'))
RL_Keras.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

inicio_Keras = time.time()

RL_Keras.fit(x=variables_train, y=tipo_train, verbose=0, epochs=100)

fin_Keras = time.time()
t_Keras = fin_Keras - inicio_Keras 
print("El modelo de Keras tarda en procesarse",t_Keras, "segundos.")
print("Se predicen un",RL_Keras.evaluate(variables_train,tipo_train)[1]* 100, "% de los casos con Keras.")
```
